```python
#Loading and Displaying Data (5 marks):
#Install necessary Python libraries for data science 

# import python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading and Displaying Data (5 marks):

# CSV file loaded using pd.read_csv()
# Excel file loaded using pd.read_excel()
# import file - csv selected
filecsv='https://omdena-datasets.s3.eu-central-1.amazonaws.com/public_datasets/%5CHealth+Conditions+and+Treatment+Dataset%5CHealth-Conditions.csv'
df_csv = pd.read_csv(filecsv, encoding = "ISO-8859-1")
df_csv.info()

# Display the first few rows of the dataset to understand its structure
df_csv.head()
```

```python
# import file - xlsx selected
filexls = 'https://omdena-datasets.s3.eu-central-1.amazonaws.com/public_datasets/%5CHomelessness+in+the+United+States+dataset+(2007-2022)%5C2007-2022-HIC-Counts-by-CoC.xlsx'
filexls = 'https://omdena-datasets.s3.eu-central-1.amazonaws.com/public_datasets/%5C2007-2022+Homeless+Populations+by+State%5C2007-2022-HIC-Counts-by-State.xlsx'
df_xls = pd.read_excel(filexls, skiprows=1)
df_xls.info()

# Display the first few rows of the dataset to understand its structure
df_xls.head()
```

```python
# Apply data manipulation techniques using Pandas

# for df_csv 
#1 select survival_rate data that is not NaN
df_csv[~df_csv['Survival_Rate'].isna()][['Survival_Rate']]

#2, apply value count on no of syptoms
print("Count condition:")
print(df_csv['condition'].value_counts(ascending=False),'\n')

print("Proportions condition :")
print(df_csv['condition'].value_counts(normalize=True, ascending=False).round(3),'\n')

# check where diagnosis based on blood samples
print(df_csv['Diagnosis'].str.contains('Blood').sum(), '\n')

# filter list of impacted records
print("Blood diagnosis:")
print(df_csv[df_csv['Diagnosis'].str.contains('Blood') == True][["condition","Diagnosis"]], '\n')

# add new feature 'Diagnosis contains Blood' to see how many sypmtons can be diagnosed with blood
df_csv['Diagnosis contains Blood']=df_csv['Diagnosis'].str.contains('Blood')

# summarize by Condition and blood diagnosis
print("Proportions Diagnosis with blood:")
print(df_csv.groupby('condition')[['Diagnosis contains Blood']].agg(['mean','sum']),'\n')

# for df_xls
print(f"Mean Total Year-Round Beds (ES, TH, SH)s: {df_xls['Total Year-Round Beds (ES, TH, SH)'].mean()} " + '\n')
```

```python
#Task 3: Statistical Analysis (10 marks)

# compute mean
print(f"mean:  {df_xls['Total Year-Round Beds (ES, TH, SH)'].mean()}" + '\n')

# compute median
print(f"median:  {df_xls['Total Year-Round Beds (ES, TH, SH)'].median()}" + '\n')

# compute standard deviation
print(f"std:  {df_xls['Total Year-Round Beds (ES, TH, SH)'].std(ddof=1)}" + '\n')
```

```python
# Task 4: Exploratory Data Analysis (15 marks)

# Data Visualization (8 marks):

# Histogram
df_xls['HMIS Participation Rate for Year-Round Beds (ES, TH, SH)'].hist()
plt.show()
```

```python
# Scatter
sns.scatterplot(data=df_xls, x='HMIS Participation Rate for Year-Round Beds (TH)', y='HMIS Participation Rate for Year-Round Beds (ES)')
plt.title("scatter plot HMIS participation rate (ES vs (TH)")
plt.show()
```

```python
# boxplot
sns.boxplot(data=df_xls, y='HMIS Participation Rate for Year-Round Beds (ES)')
plt.title("BOX plot HMIS participation rate (ES)")
plt.show()
```

```python
# outliers detected from the box plot

# calculate upper and lower limits for outlier
xmedian  = df_xls['HMIS Participation Rate for Year-Round Beds (ES)'].median()
lower_q, upper_q = df_xls['HMIS Participation Rate for Year-Round Beds (ES)'].quantile([0.25,0.75])

# interquartile range
IQR = upper_q - lower_q

# identify outliers
threshold = 1.5
lower_limit = lower_q - threshold * IQR
upper_limit = upper_q + threshold * IQR
print(f"IQR: {IQR} , lower outlier limit: {lower_limit} , upper outlier limit {upper_limit}")


#Identify outliers
outlierchk = df_xls['HMIS Participation Rate for Year-Round Beds (ES)'] 
outliers=df_xls[(outlierchk< lower_limit) | (outlierchk > upper_limit)]
print("Outliers: ")
print(outliers[['State','HMIS Participation Rate for Year-Round Beds (ES)']])

# drop rows containing outliers
df_xls.drop(outliers.index, inplace=True)
```
